```python
import nltk
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
import konlpy
from konlpy.tag import Okt   # OKt: Open Korean text
okt = Okt()
```

```python
with open('C:/ken/alice/title.txt', 'r', encoding='utf-8') as f:
    my_text = f.read()
stopwords = set(STOPWORDS)
len(stopwords)
```

```python
stopwords.add('said')
```

```python
tokens = okt.nouns(my_text)
tokens[40:50]
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

#한글폰트 사용
mpl.rc('font', family='NanumBarunpen')
mpl.rc('axes', unicode_minus=False)
```

- 아래 두 셀을 반복해서 실행하여 stop_words를 갱신하다

```python
stop_words=['따릉','뉴따릉','이용','방법','타고','매일','따릉이타','라이딩','일기','천원','더','서울','자전거','서울시','가기','형','구경','티머니','오늘','따릉이','새싹따릉']
tokens=[word for word in tokens if word not in stop_words]
tokens[:10]
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
my_title = nltk.Text(tokens, name='bicycle')
plt.figure(figsize=(15,6))
my_title.plot(50)
plt.show()
```

- 워클 그리기

```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```python
data=my_title.vocab().most_common(300)
wc=WordCloud(
    font_path='C:/windows/fonts/malgun.ttf',
    relative_scaling=0.2,
    background_color='white'
).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,6))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
