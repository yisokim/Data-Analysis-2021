```python
df = pd.read_csv('C:/ken/alice/강남C.csv')
df
```

```python
map = folium.Map(
    location=[df['위도'].mean(), df['경도'].mean()], 
    zoom_start=10
)
for i in df.index:
    folium.Marker(
        location=[df['위도'][i], df['경도'][i]],
        tooltip=df['자전거도로입구명'][i],
        popup=folium.Popup(
            df['지번 주소'][i], min_width=200, max_width=200)
    ).add_to(map)
title_html = '<h3 align="center" style="font-size:20px"><b>5개 지역 표시</b></h3>'   
map.get_root().html.add_child(folium.Element(title_html))
map
```
