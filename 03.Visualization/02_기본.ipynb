# 데이터 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 10
```

## 1. 라인 플롯

```python
plt.plot([1, 4, 9, 16])   # x = [0,1,2,3] 가정
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])   # y = x** 그래프
plt.title("y = x** 그래프")
plt.grid()
plt.show()
```

## - 스타일 지정

```python
# 스타일 문자열은 색깔(color), 마커(marker), 선 종류(line style)의 순서로 지정
plt.plot([1,2,3,4],[1,4,9,16], 'rs--')
plt.title("'rs--' 스타일")
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16], c='magenta',ls=':',lw=3,ms=10,marker='o',mec='g',mew=5,mfc='r')
plt.title("'rs--' 스타일")
plt.grid()
plt.show()
```

## - 그림 범위 지정

```python
plt.plot([1,2,3,4],[1,4,9,16], c='magenta',ls=':',lw=3,ms=10,marker='o',mec='g',mew=5,mfc='r')
plt.title("x축, y축의 범위 설정")
plt.xlim(0,5)
plt.ylim(-2,20)
plt.grid()
plt.show()
```

# 여러개의 선 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

## -복수개의 plot을 사용해 겹쳐 그리기

```python
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--')
plt.plot(t, 0.5 * t**2, 'bs:')
plt.plot(t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

## - 범례(legend)

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, label="cosine")
plt.plot(X, S, label="sine")
plt.legend(loc=2)
plt.grid()
plt.show()
```

## - x축, y축 라벨, 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("x축,y축 라벨을 표시한 플롯")
plt.plot(X, C, label="cosine")
plt.plot(X, S, label="sine")
plt.legend(loc=2)
plt.xlabel('x축 라벨'); plt.ylabel('y축 라벨')
plt.grid()
plt.show()
```
