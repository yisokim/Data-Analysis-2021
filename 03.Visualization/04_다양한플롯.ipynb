# 1. Bar chart
##  bar 명령과 barh 명령으로 바 차트(bar chart) 시각화
## 가로 방향으로 바 차트를 그리려면 barh 명령을 사용
## xerr 인수나 yerr 인수를 지정하면 에러 바(error bar)를 추가 / alpha: 투명도

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

# 타이타닉 데이터 - 선실별 승객수

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
df = titanic[['pclass','class']].groupby('class').count()
df
```

```python
y = df.pclass.values
y
```

```python
x = np.arange(len(y))
xlabel = df.index.tolist()
xlabel
```

```python
plt.title('타이타닉 선실별 승객')
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.xlabel("선실")
plt.ylabel("승객 수")
plt.show()
```

# 2. 파이차트
## 카테고리 별 값의 상대적인 비교를 해야 할 때는 pie 명령으로 파이 차트

```python
plt.axis('equal')
```

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```
