# 여친 선물 - 워드 클라우드 만들기

```python
gift_text = open('C:/ken/alice/gift.txt', encoding='utf-8').read()
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
tokens = okt.nouns(gift_text)
tokens[:10]
```

```python
# 영문자, 숫자 제거
import re
new_tokens = []
for token in tokens:
    text = re.sub('[a-zA-Z0-9]', '', token)
    if text:
        new_tokens.append(text)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 10
```

```python
import nltk
```

## - 아래 두 셀을 반복해서 실행하여 stop_words를 갱신

```python
gf_gift = nltk.Text(new_tokens, name='여친선물')
plt.figure(figsize=(15,6))
gf_gift.plot(50)
plt.show()
```

```python
stop_words = ['선물','추천','것','가격','여친','신지','몇개','투척','제품','더','남녀','이즈','버핏','빅사']
new_tokens = [word for word in new_tokens if word not in stop_words]
new_tokens[:10]
```
