```python
#00. 패키지 호출
import pandas as pd 
import numpy as np 
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.metrics import accuracy_score
```

```python
#01. 데이터셋 불러오기. 
df_감말랭이_raw = pd.read_csv("C:/ken/feature-target/감말랭이.csv")
df_감말랭이_raw.head()
```

```python
#02. 날짜 제거하기. 
df_감말랭이 = df_감말랭이_raw
del df_감말랭이['date']
df_감말랭이
```

```python
#03. 데이터셋 스케일링하기. 
scaler = StandardScaler()
df_감말랭이_scaling = scaler.fit_transform(df_감말랭이.iloc[:,:-1])
df_감말랭이_scaling = pd.DataFrame(df_감말랭이_scaling, columns=['평균기온','최저기온','최고기온','일강수량','평균이슬점온도','최소상대습도',	'평균상대습도','합계일조시간','합계일사량','평균지면온도','합계대형증발량','합계소형증발량'])

df_감말랭이_scaling['감말랭이'] = df_감말랭이.iloc[:,-1].round().astype('int')
df_감말랭이_scaling
```

```python
# 결측값 확인
df_감말랭이_scaling.isna().sum()
```

```python
#04. 데이터셋 탐색하기. 
df_감말랭이_scaling.info()
```

```python
df_감말랭이_scaling.describe()
```

```python
#05. train, test 데이터셋으로 분할하기. 
X_train, X_test, y_train, y_test = train_test_split(
    df_감말랭이_scaling.iloc[:,:-1], df_감말랭이_scaling.iloc[:,-1], test_size=0.5 
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#06. 랜덤포레스트 모델로 머신러닝하기. 
#(1) 모형 생성하기. 
rf = RandomForestClassifier(n_estimators=100,
                               max_features=6,
                               n_jobs=-1,
                               random_state=2021)

#(2) 하이퍼 파라미터 확인하기. 
RandomForestClassifier.get_params
```

```python
#(3) GridSearch를 위한 하이퍼파라미터 값 지정하기. (recycle)
params = {'max_depth':[ 3,4,5,6,7,10],
        'n_estimators':[100,200,300],
        'max_features':[6,8,10,12,14,16]}

#(4) GridsearchCV() 메서드를 이용해 훈련 모델 할당하기. 
gscv_rf = GridSearchCV(estimator=rf,
                        param_grid=params,
                        cv=5,
                        n_jobs=-1,
                        verbose=2,
                        scoring = 'accuracy')

#(5) 모형 학습하기. 
gscv_rf.fit(X_train,y_train)
```

```python
#(6) 최적 하이퍼 파라미터 확인하기. 
gscv_rf.best_params_
```

```python
#(7) 최적 파라미터에 대한 평가 점수 확인하기. 
gscv_rf.best_score_
```

```python
#(8) test 데이터셋으로 모형 예측 및 평가하기. 
gscv_rf_fit = gscv_rf.best_estimator_
rf_pred =gscv_rf_fit.predict(X_test)
r2_score(y_test,rf_pred)
```
